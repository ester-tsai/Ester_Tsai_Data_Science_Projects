## Combining all Unique IDs for the Same Business in One Cell

```python
table[table.get("Business Name") == table.iloc[384].get("Business Name")]
```

```python
for section_index, section_row in table[table.get("Business Name") == table.iloc[384].get("Business Name")].iterrows():
    print(section_row.get("EEP Unique ID"))
```

```python
for index, row in table.iterrows():
    col_with_dupe = row['dupe type']
    unique_ID_list = ""
    try:
        for section_index, section_row in table[table.get(col_with_dupe) == row.get(col_with_dupe)].iterrows():
            unique_ID = section_row.get('EEP Unique ID')
            if unique_ID is not np.nan and unique_ID not in unique_ID_list:
                unique_ID_list += unique_ID + "/"
        if len(unique_ID_list) > 0:
            unique_ID_list = unique_ID_list[:-1]
        print(unique_ID_list)
    except:
        print(index, col_with_dupe)
```
